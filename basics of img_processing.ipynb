```python
from PIL import ImageEnhance
```

```python
row=2
column=2
fig = plt.figure(figsize=(8,8))

en_img1= ImageEnhance.Color(my_image).enhance(2.5)
en_img2= ImageEnhance.Contrast(my_image).enhance(5)           #################
en_img3= ImageEnhance.Brightness(my_image).enhance(2.5)
en_img4= ImageEnhance.Sharpness(my_image).enhance(100)
fig.add_subplot(row, column, 1)
plt.imshow(en_img1)
fig.add_subplot(row, column, 2)
plt.imshow(en_img2)
fig.add_subplot(row, column, 3)
plt.imshow(en_img3)
fig.add_subplot(row, column, 4)
plt.imshow(en_img4)
```
